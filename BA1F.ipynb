```python
sk = input()
mn = 0
ans = 0
lst1 = []
lst2 = []
for i in range(len(sk)):
  if sk[i] == 'C':
    ans -= 1
  if sk[i] == 'G':
    ans += 1

  if ans < mn:
    mn = ans
  lst1.append(ans)
  lst2.append(i+1)
if mn == 0:
  print(0);
j = 0
for pos in lst1:
  if pos == mn:
    print(lst2[j])
  j+=1
```
